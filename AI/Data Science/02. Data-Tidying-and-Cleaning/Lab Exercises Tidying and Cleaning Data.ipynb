```python
import pandas as pd
import numpy as np
```

```python
temp_data = pd.read_csv('./Data/weather.csv')
temp_data.head(5)
```

```python
temp_data = temp_data.melt(id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
temp_data.head()
```

```python
temp_data.dropna(inplace=True)
temp_data.head()
```

```python
temp_data.day = pd.to_numeric(temp_data.day.apply(lambda x: x[1:]))
temp_data.head()
```

```python
temp_data['date'] = pd.to_datetime(temp_data[['year', 'month', 'day']])
temp_data.head()
```

```python
temp_data = temp_data[['id', 'date', 'element', 'temp']]
temp_data.head()
```

```python
temp_data = temp_data.pivot(index=['id', 'date'], columns='element', values='temp').reset_index()
temp_data = temp_data.rename_axis(None, axis=1)
temp_data.head()
```

```python
weather_data = pd.read_csv('https://raw.githubusercontent.com/synesthesiam/blog/master/posts/data/weather_year.csv')
```

```python
weather_data.columns = ["date", "max_temp", "mean_temp", "min_temp", "max_dew",
"mean_dew", "min_dew", "max_humidity", "mean_humidity",
"min_humidity", "max_pressure", "mean_pressure",
"min_pressure", "max_visibility", "mean_visibility",
"min_visibility", "max_wind", "mean_wind", "max_gusts",
"precipitation", "cloud_cover", "events", "wind_dir"]

weather_data.date = pd.to_datetime(weather_data.date, infer_datetime_format=True)
weather_data = weather_data.set_index('date')
weather_data.precipitation = pd.to_numeric(weather_data.precipitation.replace('T', 1e-6)).astype(float)
weather_data.events = weather_data.events.fillna('')
weather_data.head()
```

```python
weather_data.precipitation.value_counts()
```

```python
weather_data.groupby(["events"]).size().sort_values(ascending=False)
```

```python
billboard_data = pd.read_csv('Data/billboard.csv')
billboard_data['index'] = billboard_data.index
songs_data = billboard_data[['year', 'artist', 'track', 'time']]
songs_data.index.name = 'idx'
songs_data
```

```python
chart_data = billboard_data.drop(columns=['year', 'artist', 'track', 'time']).melt(id_vars=['index', 'date.entered'], var_name='week', value_name='position')
```

```python
chart_data.dropna(inplace=True)
```
